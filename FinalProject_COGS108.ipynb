# Abstract

The goal of our project was to determine if the seasons of the year influences flight behavior within the aviation industry for both pre-pandemic and pandemic years. Specifically, does the season of the year impact the number of passengers flying, the number of flights being flown, the types of flights (domestic and international) being flown, and the total load factor of flights. Through completing an exploratory analysis of our data we found that pre-COVID-19 years had a visible seasonal difference in these flight behavior components compared to COVID-19 years. In order to further investigate these differences and determine if a seasonal trend existed in both time periods, we conducted a time series decomposition, one-way ANOVA test, and Tukey HSD test on all flight behavior components of our research question. These analyses allowed us to find that the season of the year does have a statistically significant effect on the number of passengers flying, the number of flights being flown, the types of flights (domestic and international) being flown, and the total load factor of flights during pre-pandemic years. Furthermore, we were able to see that the season of the year where all of these flight components were at their highest was the summer and the season of the year where all of these components were at their lowest was the winter. Additionally, these analyses suggested that there is no significant seasonal influence on these flight components during the pandemic and thus, no relationship exists. However, we must be aware that our sample size of our COVID-19 data was much smaller in size compared to our pre-COVID-19 data and thus, this may have limited our ability in finding a statistically significant relationship for season and flight behavior during the pandemic. We hope that our findings can shed light on how seasonal patterns do exist within the aviation industry (prior to the pandemic) in order to help the aviation industry operate in a more efficient and productive way that is in accordance with the seasonal trends. Thus, this will be able to help meet and satisfy consumer demands and will also give air travel consumers the ability to better plan their trips around these seasonal patterns.

```python
# Check the types of each variable

df.dtypes
```

```python
# Convert object datatypes to floats

df['Domestic Passengers'] = df['Domestic Passengers'].replace(',', '', regex = True)
df['Domestic Passengers'] = df['Domestic Passengers'].astype(float)

df['International Passengers'] = df['International Passengers'].replace(',', '', regex = True)
df['International Passengers'] = df['International Passengers'].astype(float)

df['Total Passengers'] = df['Total Passengers'].replace(',', '', regex = True)
df['Total Passengers'] = df['Total Passengers'].astype(float)

df['Domestic Flights'] = df['Domestic Flights'].replace(',', '', regex = True)
df['Domestic Flights'] = df['Domestic Flights'].astype(float)

df['International Flights'] = df['International Flights'].replace(',', '', regex = True)
df['International Flights'] = df['International Flights'].astype(float)

df['Total Flights'] = df['Total Flights'].replace(',', '', regex = True)
df['Total Flights'] = df['Total Flights'].astype(float)

df['Domestic RPM'] = df['Domestic RPM'].replace(',', '', regex = True)
df['Domestic RPM'] = df['Domestic RPM'].astype(float)

df['International RPM'] = df['International RPM'].replace(',', '', regex = True)
df['International RPM'] = df['International RPM'].astype(float)

df['Total RPM'] = df['Total RPM'].replace(',', '', regex = True)
df['Total RPM'] = df['Total RPM'].astype(float)

# Check the type of each of our attributes again
df.dtypes
```

In order to analyze for seasonal trends within the aviation industry, we need to attribute each of our observations to a season. Based off the four seasons of a year, each month variable of our observations needs to be assigned to the season that particular month falls under.

```python
# Create a column for the season of the year

# Function to identify which month falls in which season
def Season_Column(row):
  if (row['Month'] == 12) | (row['Month'] == 1) | (row['Month'] == 2):
    return 'Winter'
  if (row['Month'] == 3) | (row['Month'] == 4) | (row['Month'] == 5):
    return 'Spring'
  if (row['Month'] == 6) | (row['Month'] == 7) | (row['Month'] == 8):
    return 'Summer'
  if (row['Month'] == 9) | (row['Month'] == 10) | (row['Month'] == 11):
    return 'Fall'

# Create a new season column by applying the above function
df = df.assign(Season = df.apply(Season_Column, axis = 1))
```

To account for any possible effects that the COVID-19 pandemic may have played in air travel behavior, we need to determine whether a given observation falls within the COVID-19 time-period. This COVID-19 time-period is based on the airline's response to the COVID-19 pandemic; canceling flights in March 2020 to reduce the spread of COVID-19 and the official declaration of the end of the COVID-19 emergency (May 2023). This will help make it easier to handle our subsequent data visualizations and analyses.

```python
# Create a column for whether the month was during the COVID-19 pandemic

# Function to indicate which year and month corresponds to a period of Covid ('Pre-Covid, During Covid, Post-Covid)
def Covid_Column(index):
    if index < 206:
        return 'Pre-COVID Period'
    elif index >= 206 and index < 244:
        return 'During COVID Period'
    elif index >= 244:
        return 'Post-COVID Period'

# Create a Covid Period column by applying the above function to the index.
df['Covid Period'] = df.reset_index().get('index').apply(Covid_Column)
```

In order to determine the typical distance that passengers are flying each month, we need to create a variable that will consider on average, how long a flight is for a given month.

```python
# Create columns for the average flight length of a month

df['Average Distance Traveled'] = df['Total RPM'] / df['Total Passengers']
df['Average Distance Traveled Domestic'] = df['Domestic RPM'] / df['Domestic Passengers']
df['Average Distance Traveled International'] = df['International RPM'] / df['International Passengers']
```

To help us more easily access and manipulate our data, we will create a new column that consists of both the year and month.

```python
# Create a new column combining year and month into the same column
df['Year-Month'] = df['Year'].astype(str) + '-' + df['Month'].astype(str)

# Reordering columns
new_column_order = ['Year', 'Month', 'Year-Month', 'Domestic Passengers',
                    'International Passengers', 'Total Passengers', 'Domestic Flights',
                    'International Flights', 'Total Flights', 'Domestic RPM', 'International RPM',
                    'Total RPM', 'Domestic Load Factor', 'International Load Factor', 'Total Load Factor',
                    'Season', 'Covid Period', 'Average Distance Traveled', 'Average Distance Traveled Domestic',
                    'Average Distance Traveled International']
df = df[new_column_order]
```

## Analysis

In order to complete the analysis of our data, we will be performing three separate analyses: time series decomposition, one-way ANOVA tests, and Tukey HSD tests. By conducting three types of analyses, this will help give us a more holistic understanding of not only whether a relationship exists between a month's season and flight behavior, but also whether a relationship exists between a month's season and the COVID-19 pandemic's flight behavior.

### Time Series Decomposition and Results

We decided on conducting time series decomposition, using the `seasonal_decompose()` function from the statsmodels library, as one of our analytical approaches because this process involves separating time series data into its main components which includes trend, seasonality, and residual. The trend component of time series decomposition indicates the long-term shifts and patterns noticed over time and displays the overall direction that the data is shifting in (i.e. if the data is increasing, the trend component will display a positive, upward slope). The seasonal component of time series decomposition indicates the repeating cyclical patterns that are seen over a certain, fixed time period and thus represents the seasonal variation of the series for a pre-determined time interval. The residual component of time series decomposition represents all the random variation that is left over once the trend and seasonal components are taken into account and have been considered for. This component portrays the error or noise in the series that can't be described by either the trend or seasonal patterns.

Through doing time series decomposition, this will provide us with the ability to determine if there are any underlying or hidden trends and patterns in the data that weren't visible by only analyzing the raw data. Through splitting up our data into the constituent parts of trend, seasonality, and residual, this will allow us to obtain a strong sense of how each of these components plays a part in the overall behavior of the data. Furthermore, by applying time series decomposition to our data, this will provide us with the ability to analyze and compare patterns across multiple time periods (i.e. pre-COVID and during COVID). This will help us observe whether different time periods have differing seasonal patterns, thus allowing us to modify our interpretations appropriately.

### One-way ANOVA and Tukey HSD Tests and Results

In order to determine whether a true, significant relationship exists within each season and specific flight behaviors (i.e. number of passengers, number of flights, the types of flight (domestic and international), and the total load factor of flights), we will conduct both a one-way ANOVA test and pairwise comparison test using TukeyHSD for each flight behavior component as this will give us a detailed account of the type of way that season impacts and interacts with these different flight behavior features. We will first use a one-way ANOVA test to assess the null hypothesis that the four seasons of the year have the same population mean for each of the flight behavior features. We decided to use this test because this analysis is particularly beneficial for data with more than two groups (unlike a t-test, which is only good for data with two groups) and unlike linear regression, ANOVA is better equipped at determining whether a relationship exists between a continuous dependent variable and one or more categorical independent variables. This analysis will allow us to determine whether a statistically significant relationship exists between the mean values of more than two groups. While the ANOVA test evaluates whether there is a significant difference at all amidst any of the seasons, it doesn't signify which season means are significantly different. In order to investigate further and in more detail of which specific seasons have statistically significant differences, pairwise comparison using TukeyHSD will be used. This will allow us to pinpoint exactly which seasons are different from one another and will do this by comparing the mean of each season to the mean of each of the other seasons. The null hypothesis for this test states that the underlying distributions of the seasons all have the same mean and with this analysis calculating the adjusted p-value for every possible pairing of seasons, we will be able to determine for each seasonal pair whether it's a significant difference. By separating our one-way ANOVA test and pairwise comparison tests by COVID time period (pre-COVID vs during-COVID), this will allow us to see if there is a significant difference in whether the season of a month affects flight behavior based on what the pandemic status is. Thus, we will be able to conclude whether our initial hypothesis of there being seasonal trends to flight behavior even holds true through the COVID-19 pandemic or if our hypothesized relationship disappears within the COVID-19 time interval.

```python
# Let's first only look at the pre-COVID data

# Extract the data for the total number of passengers for all 'Fall' seasons during the pre-covid period
test1 = pre_covid_df.loc[pre_covid_df['Season'] == 'Fall', ['Total Passengers']]

# Extract the data for the total number of passengers for all 'Winter' seasons during the pre-covid period
test2 = pre_covid_df.loc[pre_covid_df['Season'] == 'Winter', ['Total Passengers']]

# Extract the data for the total number of passengers for all 'Summer' seasons during the pre-covid period
test3 = pre_covid_df.loc[pre_covid_df['Season'] == 'Summer', ['Total Passengers']]

# Extract the data for the total number of passengers for all 'Spring' seasons during the pre-covid period
test4 = pre_covid_df.loc[pre_covid_df['Season'] == 'Spring', ['Total Passengers']]
```

```python
# Conduct a one-way ANOVA test for the total number of passengers

f_oneway(test1, test2, test3, test4)
```

Based on the results of the one-way ANOVA test, it can be seen that by using an alpha level of 0.05 the p-value is less than 0.05, thus allowing us to reject our null hypothesis that the mean total number of passengers of all four seasons are equal. We can instead accept the alternative hypothesis that states that the season of the year does have an impact on the number of passengers flying and that there is a significant difference in the total number of passengers among at least one of the four different seasons.

```python
# Conduct a pairwise comparison test for the total number of passengers

tukey = pairwise_tukeyhsd(pre_covid_df['Total Passengers'], pre_covid_df['Season'])
print(tukey)
```

While we discovered that there is an overall difference among the seasons of the year for the number of passengers flying, we still didn't know which specific seasons of the year are significantly different from each other. In order to determine how the seasons of the year differ from one another and which specific pairs of seasons are significantly different, we performed a pairwise comparison TukeyHSD test. Based on this analysis, we can see that the p-values reported were adjusted because of the multiple pairwise comparisons and this subsequently displays that among each pair of seasons, they are all significantly different from each other since all adjusted p-values are less than the alpha level of 0.05. As a result, we can reject our null hypothesis in favor of the alternative hypothesis that states that the means of all the paired seasons are not equal to each other and actually have true differences in the number of passengers flying. Therefore, we can conclude that there is a significant relationship between seasons and the number of passengers flying, the number of flights flown, the types of flights (domestic and international) flown, and the total load factor of flights pre-pandemic. Though contrary to our initial hypothesis, the season with the highest activity of these flight components is summer and the season with the lowest flight activity is winter. With our results for the COVID-19 pandemic being consistent with our initial hypothesis, we can also conclude that the COVID-19 pandemic greatly influenced flight activity as a seasonal trend could no longer be seen for any of the seasons. Thus, there is a statistically significant relationship between a month’s season and flight activity before the pandemic, but this relationship vanishes during the pandemic. However, due to the limitations of the study, further research is required to determine the precise nature of these seasonal trends, and to more fully explore the effects of pandemic-related changes on seasonal trends.
